Task - 1:
Create a Python program that converts between different units of measurement.
• The program should:
1. Prompt the user to choose the type of conversion (e.g., length, weight, volume).
2. Ask the user to input the value to be converted.
3. Perform the conversion and display the result.
4. Handle potential errors, such as invalid input or unsupported conversion types.
• Requirements:
1. Functions: Define at least one function to perform the conversion.
2. Error Handling: Use try-except blocks to handle invalid input (e.g., non-numeric values).
3. User Input: Prompt the user to select the conversion type and input the value.
4. Docstrings: Include a docstring in your function to describe its purpose, parameters, and

• Conversion Options:
1. Length:
– Convert meters (m) to feet (ft).
– Convert feet (ft) to meters (m).
2. Weight:
– Convert kilograms (kg) to pounds (lbs).
– Convert pounds (lbs) to kilograms (kg).
3. Volume:
– Convert liters (L) to gallons (gal).
– Convert gallons (gal) to liters (L).


```python
def convert_units(conversion_type, value):
    """
    Convert units based on the selected conversion type.

    Parameters:
    conversion_type (int): The type of conversion selected by the user.
    value (float): The numeric value to be converted.

    Returns:
    float: The converted value.
    """

    if conversion_type == 1:   # meters to feet
        return value * 3.28084
    elif conversion_type == 2: # feet to meters
        return value / 3.28084
    elif conversion_type == 3: # kilograms to pounds
        return value * 2.20462
    elif conversion_type == 4: # pounds to kilograms
        return value / 2.20462
    elif conversion_type == 5: # liters to gallons
        return value * 0.264172
    elif conversion_type == 6: # gallons to liters
        return value / 0.264172
    else:
        return None


def main():
    print("Unit Conversion Program")
    print("Choose the conversion type:")
    print("1. Meters (m) to Feet (ft)")
    print("2. Feet (ft) to Meters (m)")
    print("3. Kilograms (kg) to Pounds (lbs)")
    print("4. Pounds (lbs) to Kilograms (kg)")
    print("5. Liters (L) to Gallons (gal)")
    print("6. Gallons (gal) to Liters (L)")

    try:
        choice = int(input("Enter your choice (1-6): "))
        value = float(input("Enter the value to convert: "))

        result = convert_units(choice, value)

        if result is None:
            print("Error: Unsupported conversion type.")
        else:
            print("Converted value:", result)

    except ValueError:
        print("Error: Please enter a valid numeric value.")


# Run the program
main()
```

```python
def find_sum(numbers):
    """
    Calculate the sum of numbers in a list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The sum of the numbers.
    """
    return sum(numbers)


def find_average(numbers):
    """
    Calculate the average of numbers in a list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The average value of the numbers.
    """
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """
    Find the maximum number in a list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The maximum value in the list.
    """
    return max(numbers)


def find_minimum(numbers):
    """
    Find the minimum number in a list.

    Parameters:
    numbers (list): A list of numeric values.

    Returns:
    float: The minimum value in the list.
    """
    return min(numbers)


def main():
    print("Mathematical Operations on a List")
    print("1. Find Sum")
    print("2. Find Average")
    print("3. Find Maximum")
    print("4. Find Minimum")

    try:
        choice = int(input("Enter your choice (1-4): "))

        user_input = input("Enter numbers separated by spaces: ")
        numbers = [float(num) for num in user_input.split()]

        if len(numbers) == 0:
            print("Error: The list cannot be empty.")
            return

        if choice == 1:
            print("Sum:", find_sum(numbers))
        elif choice == 2:
            print("Average:", find_average(numbers))
        elif choice == 3:
            print("Maximum:", find_maximum(numbers))
        elif choice == 4:
            print("Minimum:", find_minimum(numbers))
        else:
            print("Error: Invalid operation choice.")

    except ValueError:
        print("Error: Please enter valid numeric values.")


# Run the program
main()
```

8. Nested List Average:
Write a Python function that calculates the average of all elements in a nested list.
• Requirements:
– Define a function average nested(lst) that takes a nested list lst and returns the average
of all the elements.
– Example: For the input [[1, 2], [3, 4], [5, 6]], the output should be 3.5.

```python
# 1. Flatten a Nested List
def flatten(lst):
    """
    Flattens a nested list into a single list.

    Parameters:
    lst (list): Nested list

    Returns:
    list: Flattened list
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result


# 2. Accessing Nested List Elements
def access_nested_element(lst, indices):
    """
    Access an element in a nested list using a list of indices.

    Parameters:
    lst (list): Nested list
    indices (list): List of indices

    Returns:
    element at the specified position
    """
    element = lst
    for i in indices:
        element = element[i]
    return element


# 3. Sum of All Elements in a Nested List
def sum_nested(lst):
    """
    Calculates the sum of all numbers in a nested list.

    Parameters:
    lst (list): Nested list

    Returns:
    int/float: Sum of all elements
    """
    total = 0
    for item in lst:
        if isinstance(item, list):
            total += sum_nested(item)
        else:
            total += item
    return total


# 4. Remove Specific Element from a Nested List
def remove_element(lst, elem):
    """
    Removes all occurrences of a specific element from a nested list.

    Parameters:
    lst (list): Nested list
    elem: Element to remove

    Returns:
    list: Modified list without the element
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.append(remove_element(item, elem))
        elif item != elem:
            result.append(item)
    return result


# 5. Find the Maximum Element in a Nested List
def find_max(lst):
    """
    Finds the maximum element in a nested list.

    Parameters:
    lst (list): Nested list

    Returns:
    int/float: Maximum value
    """
    maximum = float('-inf')
    for item in lst:
        if isinstance(item, list):
            maximum = max(maximum, find_max(item))
        else:
            maximum = max(maximum, item)
    return maximum


# 6. Count Occurrences of an Element
def count_occurrences(lst, elem):
    """
    Counts occurrences of an element in a nested list.

    Parameters:
    lst (list): Nested list
    elem: Element to count

    Returns:
    int: Number of occurrences
    """
    count = 0
    for item in lst:
        if isinstance(item, list):
            count += count_occurrences(item, elem)
        elif item == elem:
            count += 1
    return count


# 7. Deep Flatten
def deep_flatten(lst):
    """
    Flattens a deeply nested list into a single list.

    Parameters:
    lst (list): Deeply nested list

    Returns:
    list: Fully flattened list
    """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(deep_flatten(item))
        else:
            result.append(item)
    return result


# 8. Nested List Average
def average_nested(lst):
    """
    Calculates the average of all elements in a nested list.

    Parameters:
    lst (list): Nested list

    Returns:
    float: Average of all elements
    """
    flat = flatten(lst)
    return sum(flat) / len(flat)


# Example Testing
nested_list = [[1, 2], [3, [4, 5]], 6]

print("Flatten:", flatten([[1, 2], [3, 4], [5]]))
print("Access Element:", access_nested_element([[1,2,3],[4,5,6],[7,8,9]], [1,2]))
print("Sum:", sum_nested(nested_list))
print("Remove Element:", remove_element([[1,2],[3,2],[4,5]], 2))
print("Max:", find_max(nested_list))
print("Count:", count_occurrences([[1,2],[2,3],[2,4]], 2))
print("Deep Flatten:", deep_flatten([[[1,2],[3,4]],[[5,6],[7,8]]]))
print("Average:", average_nested([[1,2],[3,4],[5,6]]))
```
